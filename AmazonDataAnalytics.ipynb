### Data cleaning

```python
# Checking for missing values
print(df.isnull().sum())
```

```python
# droping unwanted columns
df.drop(['product_id','user_id','user_name','review_id','img_link','product_link'],axis = 1,inplace=True)
df.head()
```

```python
# droping nan columns
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df.dtypes
```

```python
# Replacing and convert into correct data type
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype('float64')
```

```python
# Replacing and convert into correct data type
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype('float64')
```

```python
# Replacing and convert into correct data type
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('int')
```

```python
# Replacing and convert into correct data type
df['rating_count'] = df['rating_count'].str.replace(',','').astype("int")
```

```python
# Replacing and convert into correct data type
df['rating'] = df['rating'].replace('|', np.nan)
df['rating'] = df['rating'].astype('float')
```

```python
df.dtypes
```

```python
df.dropna()
```

```python
df.isnull().sum()
```

### Splitting category column and creating four new categories

```python

category = df['category'].str.split('|',expand=True).rename(columns={
    0:'category_1',
    1:'category_2',
    2:'category_3',
    3:'category_4',
    4:'category_5',
    5:'category_6',
    6:'category_7',
})
category.head()
```

```python
category.info()
```

```python
category.drop(['category_5','category_6','category_7'],axis=1,inplace=True)
category.head()
```

```python
df = pd.concat([df,category],axis=1)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('category',axis=1,inplace=True)
```

```python
df.head(5)
```

### Data Analysis

 Finding Minimum, Maximum and Average rating of a product based on their category

```python
result_1 = df.groupby('category_1')['rating'].agg(["min","mean","max"])
result_1['mean'] = result_1['mean'].round(2)
result_1
```

```python
# Finding Minimum, Maximum and Average rating of a product based on their category
result_2 = df.groupby('category_2')['rating'].agg(["min","mean","max"])
result_2['mean'] = result_2['mean'].round(2)
result_2
```

```python
# Finding Minimum, Maximum and Average rating of a product based on their category
result = df.groupby('category_3')['rating'].agg(["min","mean","max"])
result['mean'] = result['mean'].round(2)
result
```

The category_3 has more categories, so this represents the product is clssified into their category so considering this as the main category

```python
plt.figure(figsize=(15,5))
plt.plot(result[['min','mean','max']],marker='o')
plt.title("ratings based on category")
plt.xlabel('categorys')
plt.ylabel('ratings')
plt.legend(['min','mean','max'],loc='lower left')
plt.xticks(rotation = 90)
plt.show()
```

Average discount of categories

```python
plt.figure(figsize=(15,8))
data = df.groupby('category_3')['discount_percentage'].agg('mean')
for i,val in enumerate(data):
    r=round(val,2)
    plt.text(r,i,str(r),ha='left',va='center')
data.plot(kind='barh')
plt.xlabel('discounts')
plt.ylabel('categories')
plt.title('Average discount of categories')
plt.show()
```

Distribution of categories

```python
# distribution of categories

plt.figure(figsize=(10, 6))
sns.countplot(x='category_3', data=df, palette='viridis')
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Top 20 category got most of the review

```python
#Top 20 category got most of the review
x=df[['category_3','rating_count']]
xx = x.sort_values(by='rating_count', ascending=False)
xx.reset_index(inplace=True,drop=True)
xx.head(20)
```

Top 20 ratings by category

```python
# top 20 ratings by category

data1=xx.loc[0:19]
plt.figure(figsize=(10, 6))
sns.barplot(x='rating_count', y='category_3', data=data1, palette='viridis',orient='h')

plt.title('Top 20 Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating Count')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Top 20 highest rated category

```python
##### Top 20 highest rated category

x=df[['category_3','rating']]
xx = x.sort_values(by='rating', ascending=False)
xx.reset_index(inplace=True,drop=True)
xx.head(20)
```

```python
data2=xx.loc[0:20]

plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='category_3', data=data2, palette='viridis',orient='h')

plt.title('Top 20 Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Total Sell of Each Category

```python
#Total Sell of Each Category

# Plotting a bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='category_3', y='mean', data=result, color='skyblue')

# Set titles and labels
plt.title('Mean Sell in Each Category')
plt.xlabel('Category')
plt.ylabel('Total Sell')

# Show the plot
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Price Summary by Category

```python
result.reset_index(inplace=True)

result_melted = result.melt(id_vars='category_3', var_name='Stat', value_name='Price')

plt.figure(figsize=(10, 6))
sns.lineplot(x='category_3', y='Price', hue='Stat', data=result_melted, marker='o')

plt.title('Price Summary by Category')
plt.xlabel('Category')
plt.ylabel('Total Sell')
plt.legend(title='Statistic')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Actual Price vs Discount Price Chart

```python
#Actual Price vs Discount Price Chart
# Plotting a bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='category_3', y='actual_price', data=df, color='skyblue', label='Actual_Price')
sns.barplot(x='category_3', y='discounted_price', data=df, color='salmon', alpha=0.7, label='Discount_Price')


# Set titles and labels
plt.title('Actual Price Vs Discount Price')
plt.xlabel('Category')
plt.ylabel('Prices')
plt.legend()

# Show the plot
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
